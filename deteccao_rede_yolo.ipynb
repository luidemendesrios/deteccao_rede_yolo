# **Importando a classe YOLO da biblioteca ultralytics**

```python
from ultralytics import YOLO #Esta classe encapsula as funcionalidades do modelo YOLO.

model = YOLO('yolo11x.pt') #Carrega um modelo YOLO pré-treinado chamado yolo11x.pt.
#O modelo é instanciado a partir da classe YOLO e será usado para realizar a detecção de objetos.
```

OBs: site do ultralytics  - https://docs.ultralytics.com/pt/tasks/detect/

Carregando imagem da web

# **Carregamento da imagem que vai ser tratada**

```python
# Utilizando o modelo é carregado para detectar objetos em uma imagem que está hospedada na web. |
#A função model() é chamada com a URL da imagem como argumento.
#O resultado é uma coleção de detecções que são armazenadas na variável results.
results = model('https://img.freepik.com/fotos-gratis/conceito-de-piquenique-com-casal_23-2147810941.jpg')

#Caso queira carregar uma imagem salva no ambiente do colab só subtituir o comando acima para o abaixo
#results = model('/content/pessoas_yolo.png')
```

```python
results = model('/content/pessoas_yolo.png') #Carregando imagem na pasta dentro do colab
```

# **Interação sobre o resultado da detecção**

Esta célula contém um loop que percorre os resultados da detecção.

**for result in results:** Itera sobre os resultados. Se você estiver processando uma única imagem, results será uma lista com um único elemento, mas se estiver processando várias imagens, cada result representará as detecções de uma imagem específica.

**boxes = result.boxes:** Esta linha extrai as informações das caixas delimitadoras detectadas na imagem. As caixas delimitadoras contêm informações sobre a localização dos objetos detectados.

**result.save(filename='result.jpg'):** Salva a imagem com as caixas delimitadoras desenhadas sobre ela em um arquivo chamado result.jpg. O método save é aplicado a cada resultado individual de detecção.

```python
for result in results:
    boxes = result.boxes
    result.save(filename='result.jpg')
```

# **Mostrar resultado da detecção na imagem**

```python
import matplotlib.pyplot as plt #Biblioteca usada para manipular dados de imagens em Python
import matplotlib.image as mpimg #Biblioteca utilizadas para exibir imagens

img = '/content/result.jpg' # define a variável img com o caminho da imagem que foi salva no Google Colab após a detecção de objetos.

#responsável por carregar e exibir a imagem processada com as detecções
image = mpimg.imread(img) #Lê a imagem processada a partir do caminho especificado na variável img e carrega-a em uma variável chamada image.
plt.imshow(image) # Exibe a imagem carregada usando a função imshow do matplotlib
plt.axis('off') #Desabilita a exibição dos eixos do gráfico para que não apareçam na imagem mostrada.
plt.show() #Inicializa a exibição da imagem no notebook Google Colab
```
